```python
a = [1, 3, 7]
b = iter(a)
print(a)
print(b)
print(next(b)) # 1
print(next(b)) # 3
print(next(b)) # 7
```

```python
a = [1, 3, 7]
for x in map(lambda x: x*x, a):
    print(x)
```

```python
r = range(2, 8)
print(len(r)) # 6
print(r[0]) # 2
print(r[5]) # 7
```
