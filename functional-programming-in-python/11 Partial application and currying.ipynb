```python
command = input("What are you doing next? ")
[action, obj] = command.split()

match command.split():
    case [action]: print(action) # interpret single-verb action
    case [action, obj]: print(obj) # interpret action, obj
```

```python
from dataclasses import dataclass, replace

@dataclass(frozen=True)
class Person:
  name: str
  age: int

p = Person("Alice", 42)
p.name  # Alice
p2 = replace(p, age=43)

print(p2)
```

```python
def maxn(n):
    def f(x):
        return max(n, x)
    return f

max0 = maxn(0)
print(max0(3)) # Equivalent to max(0, 3) -> 3
print(max0(-1)) # Equivalent to max(0, -1) -> 0
print(maxn(0)(3))
m = map(maxn(3), [1, 2, 3, 4, 5])
print(list(m))
```

```python
from functools import partial

max0 = partial(max, 0)
print(max0(3)) # 3
print(max0(-1)) # 0
m = map(partial(max, 3), [1, 2, 3, 4, 5])
print(list(m))
```

```python
def quad(a, b, c, x):
    return a*x*x + b*x + c

def quad_abc(a, b, c):
    def f(x):
        return quad(a, b, c, x)
    return f

myquad = quad_abc(1, -3, 2)
print(myquad(0)) # 2
print(myquad(1)) # 0
print(myquad(2)) # 0
```

```python
def make_print(sep):
    def f(*args):
        return print(*args, sep=sep)
    return f

print_csv = make_print(', ')
print_colon = make_print(':')
print_csv(1, 2, 3) # 1, 2, 3
print_colon(1, 2, 3) # 1:2:3

from functools import partial
print_csv = partial(print, sep=', ')
print_colon = partial(print, sep=':')
```

```python
from pymonad.tools import curry

@curry(4)
def quadc(a, b, c, x):
    return a*x*x + b*x + c

y = quadc(1, -3, 2, 0)
print(y)
f = quadc(1, -3, 2) # type: ignore
y = f(0)
print(y)
```
